# GCD(GREATEST COMMON DIVISOR) & HCF(HIGHEST COMMON FACTOR)

```python
def gcd(a, b):
    divisor = min(a, b)
    dividend = max(a, b)
    while dividend % divisor != 0:
        temp = dividend
        dividend = divisor
        divisor = temp % divisor
    return divisor

ans = gcd(12, 10)
print(ans)
```

# ARMSTRONG

```python
def armstrong(num):
    ans = 0
    k = countnumber(num)
    temp = num
    while num > 0:
        rem = num % 10
        ans = ans + rem ** k
        num = num // 10
    return temp == ans
ans = armstrong(12)
print(ans)
```

```python
def get_second_order_element(n: int, arr: [int]) -> [int]:
    l, sl, s, ss = -float('inf'), -float('inf'), float('inf'), float('inf')
    
    for item in arr:
        if item > l:
            sl = l
            l = item
        elif item > sl:
            sl = item
        if item < s:
            ss = s
            s = item
        elif item < ss:
            ss = item
    return [sl, ss]

arr = [2, 6, 4, 9, 1, 7, 5]
n = len(arr)
get_second_order_element(n, arr)
```

# MOVE ZEROES

```python
def movezeros(n: int, arr: [int]) -> [int]:
    i, nz = 0, 0
    while i < n:
        if arr[i] != 0:
            arr[nz], arr[i] = arr[i], arr[nz]
            nz += 1
        i += 1
    return arr

arr = [10, 20, 0, 2, 0, 4, 45, 99]
n = len(arr)
movezeros(n, arr)
```

# ALTERNATE NUMBER

```python
from typing import *
def alternate_number(arr: list[int]) -> list[int]:
    n = len(arr)
    ans = [0] * (n)
    pi, ni = 0, 1
    for item in arr:
        if item < 0:
            ans[ni] = item
            ni += 2
        else:
            ans[pi] = item
            pi += 2
    return ans

arr = [1, 2, -3, -1, -2, 3]
alternate_number(arr)
```

# NEXT GREATER PERMUTATION

```python
from typing import *
def reverse(arr, s, e):
    while s < e:
        arr[s], arr[0] = arr[0], arr[s]
        s += 1
        e -= 1
def next_greater_permutation(arr: list[int]) -> list[int]:
    n = len(arr)
    i = n - 1
    while i > 0 and arr[i - 1] >= arr[i]:
        i -= 1
    bridge, k = i - 1, i
    if k == 0:
        reverse(arr, 0, n - 1)
        return arr
    i = n - 1
    while arr[i] <= arr[bridge]:
        i -= 1
    arr[i], arr[bridge] = arr[bridge], arr[i]
    reverse(arr, k, n-1)
    return arr

arr = [7, 2, 9, 8, 6]
next_greater_permutation(arr)
```

# BINARY SEARCH

```python
def search(arr: [int], target:int):
    n = len(arr)
    s, e = 0, n - 1
    while s <= e:
        mid = (s + e) // 2
        if target < arr[mid]:
            e = mid - 1
        elif target > arr[mid]:
            s = mid + 1
        else:
            return mid
    return -1
# n = int(input())
arr = list(map(int, input('enter your array: ', ).strip().split(',')))
target = int(input('enter your target value: ', ))
search(arr, target)
```

# INSERT POSITION

```python
def search_insert(arr: [int], target: int) -> int:
    n = len(arr)
    if n == 0:
        return 0
    ans = n
    s, e = 0, n - 1
    while s <= e:
        mid = (s + e) // 2
        if arr[mid] < target:
            s = mid + 1
        else:
            e = mid - 1
            ans = mid
    return ans

arr = [10, 20, 30, 50, 60]
target = 40
search_insert(arr, target)
```

# SQUARE ROOT

```python
def flootsqrt(n):
    s, e = 1, n
    ans = 1
    while s<=e:
        mid = s + (e - s) // 2
        if mid <= n / mid:
            ans = mid
            s = mid + 1
        else:
            e = mid - 1
    return ans

n = int(input())
flootsqrt(n)
```

# XOR Queries of a subarray

```python
def xorQueries(arr: list[int], queries: list[list[int]]) -> list[int]:
    n = len(arr)
    prefix = [0] * (n)
    prefix[0] = arr[0]
    for i in range(1, n):
        prefix[i] = prefix[i-1] ^ arr[i]
    ans = []
    for l, r in queries:
        if l == 0: temp = prefix[r]
        else: temp = prefix[r] ^ prefix[l-1]
        ans.append(temp)
    return ans

arr = [1,3,4,8]
queries = [[0,1], [1,2], [0,3], [3,3]]
xorQueries(arr, queries)
```

# Total Sum XOR Pairs

```python
def sumxor(arr, n):
    bitarr = [0] * (32)
    for i in range(32):
        cnt = 0
        for item in arr:
            if item & (1 << i): cnt += 1
        bitarr[i] = cnt
    ans = 0
    for i in range(32):
        n1 = bitarr[i]
        n0 = n - bitarr[i]
        temp = n0 * n1
        ans = ans + (1 << i) * temp
    return ans

arr = [7, 3, 5]
n = len(arr)
sumxor(arr, n)
```

# Finding a Missing and a Repeating Number

```python
def findTwoElement(arr, n):
    res = 0
    for i in range(1, n + 1):
        res = res ^ i
    for item in arr:
        res = res ^ item
    rsBit = res & (~(res-1))
    left = 0
    right = 0
    for i in range(1, n+1):
        if i & (rsBit): right = right ^ i
        else: left = left ^ i
    for item in arr:
        if item & (rsBit): right = right ^ item
        else: left = left ^ item
    missing = 0
    repeating = 0
    if left in arr:
        repeating = left
        missing = right
    else:
        repeating = right
        missing = left
    return [repeating, missing]

n = 2
arr = [2, 2]
findTwoElement(arr, n)
```

```python
def majorityelement(nums: list[int]) -> int:
    dp = {}
    n = len(nums)
    for item in nums:
        dp[item] = dp.get(item, 0) + 1
    ans = None
    for k, v in dp.items():
        if v > n // 2:
            ans = k
            break
    return ans

nums = [3, 2, 3]
majorityelement(nums)
```

```python
# ROTATE IMAGE
def rotate(arr: list[list[int]]) -> None:
    n = len(arr)
    for i in range(n):
        for j in range(n):
            if i > j:
                arr[i][j], arr[j][i] = arr[j][i], arr[i][j]
    for i in range(n):
        s = 0
        e = n - 1
        while s <= 0:
            arr[i][s], arr[j][e] = arr[j][e], arr[i][s]
            s += 1
            e -= 1
    return arr

arr = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
rotate(arr)
```

Given an array of integers NUMS and an integer K return the total number of subarrays whose num equals ti k.
A sub array is contigous non-empty sequnece of elements with an array
Input = nums = [1, 1], k=2
output = 2

```python
def subarraysum(arr: list[int], k:int) -> int:
    n = len(arr)
    ans = 0
    prefix = [0] * (n)
    dp = {0:1}
    for i in range(n):
        if i == 0:
            prefix[i] = arr[i]
        else:
            prefix[i] = prefix[i - 1] + arr[i]
        rem = prefix[i] - k
        ans = ans + dp.get(rem, 0)
        dp[prefix[i]] = dp.get(prefix[i], 0) + 1
    return ans

arr = [1, 1]
k = 2
subarraysum(arr, k)
```

```python
def set_zeroes(arr: list[list[int]]) -> None:
    col0 = False
    row0 = False
    m = len(arr)
    n = len(arr[0])
    for j in range(n):
        if arr[0][j] == 0:
            row0 = True
            break
    for i in range(m):
        if arr[i][0] == 0:
            col0 = True
            break
    for i in range(1, m):
        for j in range(1, n):
            if arr[i][j] == 0:
                arr[i][0] = 0
    for i in range(1, m):
        for j in range(1, n):
            if arr[i][0] == 0 or arr[0][j] == 0:
                arr[i][j] = 0
            if row0 == True:
                for j in range(n):
                    arr[0][j] == 0
            if col0 == True:
                for i in range(m):
                    arr[i][0] = 0
    return arr

arr = [[1,1,1], [1,0,1], [1,1,1]]
set_zeroes(arr)
```

```python
from typing import *

def atmostkchar(name: str, k: int) -> int:
    if k <= 0: return 0
    n = len(name)
    s, e = 0, 0
    dp = {}
    ans = 0
    
    while e < n:
        # FIX: Use name[e] instead of name[0]
        ch = name[e]
        dp[ch] = dp.get(ch, 0) + 1
        
        # Shrink window if unique characters > k
        while len(dp) > k:
            left_ch = name[s]
            dp[left_ch] -= 1
            if dp[left_ch] == 0:
                del dp[left_ch]
            s += 1
        
        # The number of substrings ending at 'e' with AT MOST k distinct chars
        ans += (e - s + 1)
        e += 1
        
    return ans

def countsubstring(name: str, k: int):
    # Standard inclusion-exclusion for "exactly k"
    return atmostkchar(name, k) - atmostkchar(name, k - 1)

name = 'aacfssa'
k = 3
print(countsubstring(name, k)) # Output: 5
```

```python
def is_valid(dp, checker):
    # Check if every character required by 'checker' is present in 'dp' with enough frequency
    for k, v in checker.items():
        if dp.get(k, 0) < v:
            return False
    return True

def min_window(name: str, t: str) -> str:
    s, e = 0, 0
    n = len(name)
    
    checker = {}
    for ch in t:
        checker[ch] = checker.get(ch, 0) + 1
        
    ansS, ansE = -1, float('inf')
    dp = {}
    
    while e < n:
        # 1. Expand the window by adding the character at 'e'
        ch_e = name[e]
        dp[ch_e] = dp.get(ch_e, 0) + 1
        
        # 2. If the window is valid, try to shrink it from the left (s)
        while is_valid(dp, checker):
            # Update the best answer if the current window is smaller
            if (e - s) < (ansE - ansS):
                ansS, ansE = s, e
            
            # Remove character at 's' and move the left pointer
            ch_s = name[s]
            dp[ch_s] -= 1
            if dp[ch_s] == 0:
                del dp[ch_s]
            s += 1
        
        # 3. Always move the right pointer forward
        e += 1
            
    return "" if ansS == -1 else name[ansS: ansE + 1]

name = 'ADDBECODEBANC'
t = 'ABC'
print(min_window(name, t))
```
